Le graphique étant illisible, je le coupe en 2 morceaux.

```python
n = len(data_2020)
taille_morceau = n // 2 + (n % 2 > 0)

for i in range(0, n, taille_morceau):
    subset = data_2020.iloc[i:i+taille_morceau]
    plt.figure(figsize=(6, 4))
    sns.barplot(data=subset, x='Country Name', y='Moyenne d\'années de scolarité, Total')
    plt.xticks(rotation=45, ha='right')
    plt.title(f'Graphique pour les pays {i+1} à {min(i+taille_morceau, n)}')
    plt.tight_layout()
    plt.show()
```
